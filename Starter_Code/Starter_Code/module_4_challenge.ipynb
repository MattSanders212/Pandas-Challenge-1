## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
transform_df = pd.DataFrame(df)
subtotal_calc = transform_df[["unit_price", "qty"]].copy()
subtotal_calc["line_subtotal"] = subtotal_calc["unit_price"] * subtotal_calc["qty"]
subtotal_calc
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
data = pd.DataFrame(df)
shipping_info = data[["first", "last","client_id", "order_id", "unit_price", "unit_cost", "unit_weight", "qty",]].copy()
def shipping_weight(row):
        total_weight = row["unit_weight"] * row["qty"]
        if total_weight > 50:
            return 7 * total_weight
        elif total_weight <= 50:
            return 10 * total_weight

shipping_info["shipping_price"] = shipping_info.apply(shipping_weight, axis = 1)
shipping_info
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
data = shipping_info.copy()
#def total_price_per_unit(row):
    #sub_total = ((row["unit_price"]) + (row["shipping_price"]/row["qty"])) 
    #sales_tax = sub_total * (9.25/100)
    #total_price_unit = sub_total + sales_tax
    #return total_price_unit
#shipping_info["Total_price_per_unit"] = shipping_info.apply(total_price_per_unit, axis = 1)
def total_price(row):
    sub_total_tot = ((row["unit_price"] * row["qty"]) + row["shipping_price"]) 
    sales_tax = sub_total_tot * (9.25/100)
    total_p = sub_total_tot + sales_tax
    return total_p
shipping_info["total_price"] = shipping_info.apply(total_price, axis = 1)
shipping_info
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
data = shipping_info.copy()
def total_cost(row):
    cost =  (row["unit_cost"] * row["qty"]) + row["shipping_price"]
    return cost
shipping_info["cost"] = shipping_info.apply(total_cost, axis = 1)
shipping_cost = shipping_info
shipping_cost 
```

```python
# Create a column for the profit of
# each line using line cost and line price
shipping_cost["Profit"] = shipping_cost["total_price"] - shipping_cost["cost"]
shipping_cost
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
#33615: $8377308.52
#66037: $10259514.79
#46820: $9743794.36
#38378: $12906550.87
#24741: $82268892.02
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
top_5_clients = [33615, 66037, 46820, 38378, 24741]
client_33615 = {33615: "total_price"}
client_66037 = {66037: "total_price"}
client_46820 = {46820: "total_price"}
client_38378 = {38378: "total_price"}
client_24741 = {24741: "total_price"}
for client in top_5_clients:
   total_spent = shipping_cost.loc[shipping_cost["client_id"] == client, 'total_price'].sum()
   if client == 33615:
      client_33615[client] = total_spent
   elif client == 66037:
      client_66037[client] = total_spent
   elif client == 46820:
      client_46820[client] = total_spent
   elif client == 38378:
      client_38378[client] = total_spent
   elif client == 24741:
      client_24741[client] = total_spent
print(client_33615)
print(client_66037)
print(client_46820)
print(client_38378)
print(client_24741)
    #if shipping_cost[shipping_cost["client_id"].isin(top_5_clients)] == True:
   # filtered_totals = shipping_cost[shipping_cost["client_id"].isin(top_5_clients)]
   # filtered_totals.head(30)



```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
summary_data_frame = shipping_cost[["client_id", "qty", "shipping_price", "total_price", "cost", "Profit"]].copy()

summary_pt_2 = summary_data_frame.loc[summary_data_frame["client_id"].isin(top_5_clients)]
final_form = summary_pt_2.pivot_table(index = "client_id", aggfunc="sum").reset_index().sort_values(by = "Profit", ascending = False)
final_form
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
updated_names = final_form.rename(columns = {"client_id" : "Client ID", "Profit": "Total Profit", "cost": "Total Cost", "qty":"Units", "shipping_price":"Shipping", "total_price":"Total Revenue"})
updated_order = updated_names[["Client ID", "Units", "Shipping", "Total Revenue", "Total Cost", "Total Profit"]]
mil_scale =["Shipping", "Total Revenue", "Total Cost", "Total Profit"]
def scale_to_millions(row, mil_scale):
        return f"${(row[mil_scale]/1_000_000).round(2)}M"
data_formated = updated_order.copy()
for columns in mil_scale:
    data_formated[columns]= updated_order.apply(scale_to_millions, axis = 1, mil_scale = columns)
data_formated
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
sorted_order =  data_formated.sort_values("Total Profit", ascending = False)
sorted_order
```
